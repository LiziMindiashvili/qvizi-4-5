```python
conn = sqlite3.connect("liziii.sqlite3")
cursor = conn.cursor()

company = input("შეიყვანეთ კომპანია: ")
rating = float(input("შეიყვანეთ რეიტინგი: "))

cursor.execute("INSERT INTO lizii (REF, Rating) VALUES (?, ?)", (company, rating))
conn.commit()
```

```python
import sqlite3
import matplotlib.pyplot as plt


conn = sqlite3.connect("liziii.sqlite3")
cursor = conn.cursor()


cursor.execute("""
    SELECT substr(Rating, 1, 4) as Year, COUNT(*) 
    FROM lizii 
    GROUP BY Year
    ORDER BY Year
""")
data = cursor.fetchall()
conn.close()


years = [row[0] for row in data]
counts = [row[1] for row in data]


plt.figure(figsize=(8, 8))
plt.pie(counts, labels=years, autopct='%1.1f%%', startangle=90)
plt.title("წრიული დიაგრამა: ჩანაწერების რაოდენობა წლების მიხედვით")
plt.show()

import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("liziii.sqlite3")
cursor = conn.cursor()


cursor.execute("SELECT Rating, COUNT(*) FROM lizii GROUP BY Rating ORDER BY Rating")
data = cursor.fetchall()
conn.close()

ratings = [row[0] for row in data]
counts = [row[1] for row in data]

plt.figure(figsize=(10, 5))
plt.bar(ratings, counts, color='purple')
plt.title("ბარის დიაგრამა: რეიტინგების რაოდენობა")
plt.xlabel("რეიტინგი")
plt.ylabel("ჩანაწერების რაოდენობა")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("liziii.sqlite3")
cursor = conn.cursor()

cursor.execute("SELECT Rating FROM lizii")
ratings = [row[0] for row in cursor.fetchall()]
conn.close()

plt.figure(figsize=(8, 5))
plt.hist(ratings, bins=10, color='green', edgecolor='black')
plt.title("ჰისტოგრამა: რეიტინგების განაწილება")
plt.xlabel("რეიტინგი")
plt.ylabel("რაოდენობა")
plt.tight_layout()
plt.show()
```
